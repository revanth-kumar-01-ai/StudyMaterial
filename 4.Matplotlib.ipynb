```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
xData = np.random.random(50) * 100
yData = np.random.random(50) * 100
```

# Scatter plot 
<p>
   A scatter plot used to visualisation the relationship between two numerical variables each point scatter plot represents an observation. with its position on the x-axis representing one variable and its position on the y-axis representing the other.
</p>

```python
scatter = plt.scatter(xData, yData, c='r', marker="*", s=150, alpha=0.3)
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.figure(figsize= (0.2, 0.8))
plt.show()
```

# Line plots 
<p>
    Use a line plot to visualize trends or changes in a continuous variable over time or a sequence.
</p>

```python
year = list(range(2006, 2022))
weight = np.random.randint(80, 90, size = 16)
```

```python
plt.plot(year, weight, 'r--', lw = 2)
```

# Bar plots 
<p>
    A bar plot (or bar chart) is used to represent categorical data with rectangular bars. The height (or length) of each bar is proportional to the value it represents. Bar plots are ideal for comparing different categories or groups.
</p>
<div>
   <h3> When to Use a Bar Plot:</h3>
To compare the sizes or frequencies of different categories.<br/>
To show the distribution of categorical data.<br/>
To compare parts of a whole when using stacked bar plots.<br/>
</div>

```python
programingLanguage = ['c', 'c++', 'go', 'python', 'java']
UsagePercentage = [74, 80, 66, 90, 100]

plt.bar(programingLanguage, UsagePercentage, color = 'r', align = "edge", width = 0.1, edgecolor = 'blue', lw = 2)
plt.show()
```

# Histogram 
<p>
    A histogram is a type of plot that represents the distribution of a dataset. It is similar to a bar plot but is used for continuous data rather than categorical data. A histogram divides the range of data into intervals (called "bins") and counts how many data points fall into each bin.
</p>
<ul>
    <h3>When to Use a Histogram</h3>
    <li>To visualize the distribution of a single continuous variable.
</li>
    <li>To understand the spread, central tendency, and variability of the data.
</li>
    <li>To identify patterns such as skewness, multimodality, or the presence of outliers.
</li>
</ul>

```python
ages = np.random.normal(20, scale = 1.5, size = 1000)
```

```python
plt.hist(ages, bins=20)
plt.show()
```

# Pie chart 
<p>
    A pie chart is a circular statistical plot used to represent proportions or percentages of a whole, with each slice corresponding to a category.
</p>
<div>
    <ul>
        <h3>When to Use a Pie Chart</h3>
<li>To show the percentage distribution of a single categorical variable.</li>
<li>To illustrate the proportion of different categories within a whole.</li>
<li>To provide a visual comparison of parts to a whole when there are a few categories.</li>
    </ul>
</div>

```python
programingLanguage = ['c', 'c++', 'go', 'python', 'java']
UsagePercentage = [74, 80, 66, 90, 100]

explodes = [0, 0, 0, 0, 0.2]

plt.pie(UsagePercentage,
       labels= programingLanguage,
       explode = explodes,
       autopct="%.2f", 
       pctdistance= 0.7,
       startangle=90)
plt.show()
```

# Box blot 
<p>
    A box plot (or box-and-whisker plot) is a graphical representation of the distribution of a dataset. It displays the median, quartiles, and potential outliers of the data. Box plots are useful for visualizing the spread and skewness of the data, as well as identifying any outliers.
</p>
<ul>
<h3> When to Use a Box Plot:</h3>
<li> To visualize the distribution and spread of data.</li>
<li> To compare distributions between multiple groups or datasets.</li>
<li> To identify outliers and understand data skewness.</li>
</ul>

```python
heights = np.random.normal(172, scale = 8, size = 300)
plt.boxplot(heights, patch_artist=True)
plt.show()
```

# multiple figuers 

```python
x1, y1 = np.random.random(100), np.random.random(100)
x2, y2 = np.arange(100), np.random.random(100)
```

```python
plt.figure(1)
plt.scatter(x1, y1) 

plt.figure(2)
plt.plot(x2, y2) 

plt.show()
```

```python
x = np.arange(100) 

fig, axis = plt.subplots(2, 2) 

axis[0, 0].plot(x, np.sin(x))
axis[0, 0].set_title("sine wave")

axis[0, 1].plot(x, np.cos(x))
axis[0, 1].set_title("Cosine wave")

axis[1, 0].plot(x, np.random.random(100))
axis[1, 0].set_title("random function")

axis[1, 1].plot(x, np.log(x+1))
axis[1, 1].set_title("log function") 
axis[1, 1].set_xlabel("Test")

fig.suptitle("four plots")

# Adjust layout to prevent overlap
plt.tight_layout()

plt.savefig("fourPlots.png", dpi = 300, transparent= False)

plt.show()
```

# 3d plot
